```python
import pandas as pd 
df = pd.read_csv(r'./data/hh_demographic.csv')
df.head()
```

```python
num_list = [i for i in range(1,2501)]
len(num_list)
```

```python
df2 = pd.DataFrame()
df2['household_key'] = num_list
df2.head()
```

```python
df3 = pd.merge(df,df2,how='outer', on = 'household_key').sort_values(by=['household_key'])
df3
```

```python
df3.to_csv('./data/hh_demographic_filling_missing_data.csv',index=False)
```
